```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
unos_df = pd.read_csv('data/unos_waitlist_071424.csv', header = [0,1])
unos_df.head()
```

```python
unos_df.columns = [c[0] + "_" + c[1] for c in unos_df.columns]
unos_df.head()
```

```python
print('Rows: ', unos_df.shape[0])
print('Columns: ', unos_df.shape[1])
```

```python
# All columns except Organ_type should be int64

cols = [i for i in unos_df.columns if i not in ['Organ_Type', '< 1 Year_All Genders', '< 1 Year_Male',
                                          '1-5 Years_All Genders', '1-5 Years_Male', '1-5 Years_Female',
                                          '6-10 Years_All Genders', '6-10 Years_Male', '6-10 Years_Female',
                                          '11-17 Years_Male', '11-17 Years_Female']]

unos_df[cols] = unos_df[cols].replace({',': ''}, regex=True)

for col in cols:
    unos_df[col]=pd.to_numeric(unos_df[col])
    
unos_df.info()
```
